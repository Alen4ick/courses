 Префиксная сумма является очень полезным инструментом для аналитики, в частности для определения кол-ва событий на каком-то промежутке.
 С помощью префиксной суммы удобно отвечать на запрос суммы на отрезке [L, R), что будет являться, prefixsum[R] - prefixsum[L]. Соответственно, если нам нужно посчитать сумму с включенными границами, то надо увеличить на единицу R, в списке префиксных сумм.

```python
nums = [int(x) for x in input().split()]
def prefix_sum(nums):
    n = len(nums)
    prefixsum = [0] * (n + 1)
    for i in range(1, n + 1):
        prefixsum[i] = prefixsum[i - 1] + nums[i - 1]
    return prefixsum
# Стоит отметить, что берется запрос на [L, R)
def sum_l_r(l, r,prf = prefix_sum(nums)):
    return prf[r] - prf[l]
s_l_r = sum_l_r(3, 6)
s_l_r
```

Стоит отметить, что вызов функции в Python делает не inline, поэтому при нахождении суммы на отрезке [L, R) лучше писать явно prefisum(r) - prefixsum(l), а не вызывать функцию.

*_Решение._*

```python
nums = [int(x) for x in input().split()]
l, r = map(int, input().split())
def prexsum(lst):
    n = len(lst)
    prefixsum = [0] * (n + 1)
    for i in range(1, n + 1):
        prefixsum[i] = prefixsum[i - 1] + 1 if lst[i - 1] == 0 else prefixsum[i-1]
    return prefixsum
prefsum = prexsum(nums)
print(prefsum)
print(prefsum[r] - prefsum[l])
```

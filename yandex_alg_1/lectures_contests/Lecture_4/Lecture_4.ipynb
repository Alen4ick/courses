# Что такое сортировка подсчетом?
## _Пусть необходимо отсортировать массив из N целых чисел, каждое от 0 до K. (Желательно, чтобы числа были небольшие, чтобы не создавать большой доп массив)_
## Обычная сортировка займет O(NlogN)
## _Решение_
## Будем считать количество вхождений каждого числа, а затем выводить каждое число столько раз, сколько оно встречалось. Это займет O(N+K) и O(K) дополнительной памяти.
### _Почему именно столько как описано сверху_?
### Потому что мы при заполнении нашего нового списка из K элементов делаем операции прибавления 1 в ячейку списка, всего таких операций N штук, т.к, исходно дается список N элементов, а каждая операция это O(1), следовательно, O(N). Потом нам надо еще создать список из K элементов O(K). Ну и соответственно на этот же список выделяет O(K) доп. памяти
## Интервал значений можно сдвинуть, чтобы он не был от 0 до K, а от минимального до максимального значения в массиве.

```python
lst = [int(x) for x in input().split()]
max_ = max(lst)
min_ = min(lst)
k = max_- min_ + 1 # Длина доп массива
dop_m = [0] * k
for el in lst:
    dop_m[el - min_]+=1
ind = 0
for j in range(k):
    for i in range(dop_m[j]):
        lst[ind] = min_ + j
        ind+=1
print(lst)
```

# Задача. Даны 2 положительных числа X, Y. Можно ли получить 1 число перестановкой цифр другого.
## _Решение. Создадим для каждого числа список длиной 10, где в каждой ячейке будем хранить сколько раз встречалась такая цифра в записи числа. Ну и в конце собственно сравним два этих списка на совпадение.

```python
def check(x, y):
    def digits_lst(num):
        num_lst = [0] * 10
        while num >0:
            last = num % 10
            num_lst[last] +=1
            num//=10
        return num_lst
    digitsx = digits_lst(x)
    digitsy = digits_lst(y)
    for i in range(10):
        if digitsx[i] != digitsy[i]:
            return False
    return True
print(check(int(input()), int(input())))
            
```

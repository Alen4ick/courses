### Найтим максимум и предмаксимум в последовательности

## Решение. Заведем две переменные для первого и второго максимума. Возьмум первые два числа из последовательности и запишем большее из них в переменную для первого максимума, а меньше - для второго. Пройдем по последовательности. Если очередное число больше первого максимума, то запишем во второй максимум значение первого, а в первый - текущее число. Если только больше второго, запишем текущее число во второй максимум.

```python
def findmax2(seq):
    max_ = max(seq[0], seq[1])
    predmax = min(seq[0], seq[1])
    for i in range(2, len(seq)):
        if seq[i] > max_:
            predmax = max_
            max_ = seq[i]
        elif seq[i] > predmax:
            predmax = seq[i]
    return max_, predmax
```

## Дана последовательность слов. Вывести все самые короткие слова через пробел

```python
def shortwords(words):
    min_len = len(words[0])
    for i in range(len(words)):
        if len(words[i]) < min_len:
            min_len = len(words[i])
    ans = []
    for word in words:
        if len(word) == min_len:
            ans.append(word)
    return ' '.join(ans)
print(shortwords(['13','5','1312','933','35', '1']))
```

```python
def count_water(h):
    ind_max = 0
    for i in range(len(h)):
        if h[i] > h[ind_max]:
            ind_max = i
    ans = 0
    max_temp = h[0]
    for i in range(1, ind_max):
        if h[i] > max_temp:
            max_temp  = h[i]
        ans += max_temp - h[i]
    max_temp = h[len(h) - 1]
    for i in range(len(h) - 1, ind_max, -1):
        if h[i] > max_temp:
            max_temp  = h[i]
        ans += max_temp - h[i]
    return ans
print(count_water([3, 1, 4, 3, 5, 1, 1, 3, 1]))
        
        
        
        
```

## Упрощенное решение

```python
def easy_one(str_):
    last = str_[0]
    ans = []
    for letter in str_:
        if letter != last:
            ans.append(last)
            last = letter
    ans.append(last)
    return ' '.join(ans)
print(easy_one('AAABBBSBD;KJFSD;KG'))
```

```python
def rle (str_):
    def pack(str_, count):
        if count > 1:
            return str_ + str(count)
        return str_
    last_let = str_[0]
    ans = []
    last_count = 0
    for i in range(len(str_)):
        if str_[i] != last_let:
            ans.append(pack(last_let, i - last_count))
            last_count = i
            last_let = str_[i]
    ans.append(pack(str_[last_count], len(str_) - last_count))
    return ''.join(ans)
print(rle('BBBBBBBB'))
    
```
